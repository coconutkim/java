# 멀티 스레드

- 운영체제는 대부분 멀티 태스킹을 한다
- 운영체제의 페이지 교체 기법, 멀티 스레드, 멀티 태스킹


- 채팅방을 일종의 프로세스로 본다
- 스레드는 1개가 아니고 여러개다
- 여러개가 동시에 참여하면 다중 스레드
- 스레드 thread는 코드의 실행 흐름을 말한다
- 스레드가 많이 처리할수록 cpu는 좋다?
- cpu는 1초만 간격을 줘도 긴 시간이다
- 스레드는 현재 실행 중인 프로그램 안의 단위가 된다?
- 프로세스가 실행 중이면 안에서 동작을 해야 하는데 그게 메소드?

- 보통 c 드라이브는 운영체제가 있는 곳이다
- 데이터는 주로 d 드라이브에 간다


- 보통의 기본 자바 코드는 싱글 스레드
- 멀티의 경우 실행 중인 스레드가 하나라도 있다면 프로세스는 종료되지 않는다


- runnable이라는 인터페이스
- 메인과 관계없이 스레드가 run()을 호출한다?




- 처음엔 생산자와 소비자 둘 다 wait
- 예를 들어 재고가 다 떨어질 때까지 생산자가 wait
- 이후 재고가 떨어지면 소비자가 생상자에게 notify
- 생산자가 다시 생산할 때까지 소비자는 wait
- 생산자는 빵을 다 만들면 소비자에게 notify



- 다른 스레드가 종료될 때까지 기다렸다가 실행을 해야 한다 >> join

- yield는 자신을 중지시키고(대기 상태) 다른 스레드가 실행이 된다
- 다른 스레드에게 제어권을 넘기는 것


- 자원 공유
- 동기화를 걸면 다른 스레드가 변경할 수 없도록 객체에 잠금이 걸린다
- 예를 들어 입금하기 전까지 출금이 안 된다
