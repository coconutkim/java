`BufferedReader`의 주된 역할은 입력을 보다 효율적으로 읽기 위한 기능 개선입니다.

1. **버퍼링**: 
   - `BufferedReader`는 내부적으로 버퍼를 사용하여 데이터를 한 번에 더 많은 양을 읽습니다. 이는 작은 단위로 여러 번 읽는 것보다 성능이 향상됩니다.

2. **줄 단위 읽기**: 
   - `readLine()` 메서드를 사용하면 한 줄씩 간편하게 읽을 수 있습니다. 이를 통해 사용자는 줄 단위로 입력된 텍스트를 쉽게 처리할 수 있습니다.

3. **문자 집합 처리**:
   - `InputStreamReader`와 함께 사용하여 바이트 스트림을 문자 스트림으로 변환할 수 있습니다. 이는 입력 데이터가 다양한 문자 인코딩을 지원할 수 있게 합니다.

결론적으로, `BufferedReader`는 특별한 역할을 수행하는 것은 아니지만, 효율성과 사용 편의성을 높이는 데 기여합니다. Java에서 입력을 처리할 때 성능을 고려해야 할 경우, `BufferedReader`는 매우 유용한 도구입니다.

### JAVA에서의 추상화 기법

특정 클래스를 상속받은 경우,
부모의 특정 메서드들을 무조건 재정의하도록 강제하는 기법이다.

특정 메서드를 재정의하도록 강제함으로써, 자식 클래스들을 작성하기 위한
가이드의 역할을 할 수 있다.

즉, 추상화 기법은 java 클래스를 작성하기 위한 설계도를 소스코드 형태로
제시하는 역할을 한다.
