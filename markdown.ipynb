Visual Studio (VS)에서 마크다운 문서를 작성할 때 사용할 수 있는 마크다운 표기법은 다음과 같습니다:

### 1. 제목(Heading)
- `# 제목 1`
- `## 제목 2`
- `### 제목 3`

### 2. 강조(Emphasis)
- **굵게**: `**텍스트**` 또는 `__텍스트__`
- *기울임*: `*텍스트*` 또는 `_텍스트_`
- ~~취소선~~: `~~텍스트~~`

### 3. 목록(Lists)
- 순서 없는 목록:
  - `- 항목`
  - `* 항목`
- 순서 있는 목록:
  - `1. 항목`
  - `2. 항목`

### 4. 링크(Links)
- `[링크 텍스트](URL)`

### 5. 이미지(Images)
- `![대체 텍스트](이미지 URL)`

### 6. 코드(Code)
- 인라인 코드: `` `코드` ``
- 블록 코드:
  ````
  ```
  코드 내용
  ```
  ````

### 7. 인용(Blockquote)
- `> 인용문`

### 8. 수평선(Horizontal Rule)
- `---`, `***`, 또는 `___`

### 9. 테이블(Tables)
```
| 헤더1 | 헤더2 |
|-------|-------|
| 내용1 | 내용2 |
```

### 10. 체크박스(Task List)
- `[ ] 할 일`
- `[x] 완료된 항목`
