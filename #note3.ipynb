>로컬 클래스
- 메소드 내부에서 선언되는 중첩 클래스


- 사용자가 잘못 입력한 것을 예외시킨다 >> 사용자 정의 예외


- can't use null >> NullPointerException
- available in java 17 versions and above

- 파일: 운영체제가 읽을 수 있는 하나의 단위
- 작업 중인 파일을 건드리면 문제가 생긴다

- finally 블록에서 예외 발생 여부와 상관없이 안전하게 닫는다

### 예외 떠넘기기
- 메소드를 호출한 곳으로 예외를 떠넘기는 것
- 떠넘긴걸 받았으면 처리를 해야 한다
- thorws 키워드 활용 / 여러개가 나오면 콤마로 이어준다
- 예외를 떠넘길 곳에서 키워드를 작성한다
- 예를 들어 thorws classnotfoundexception
- 떠넘길 곳이 없으면 main에 맡긴다

### 사용자 정의 예외
- 예를 들어 잔고 부족이라는 예외는 표준 라이브러리에 없다
- 그럼 만들면 된다
- 컴파일러가 체크하는 일반 예외로 하거나, 컴파일러가 체크하지 않는 실행 예외도 선언 가능
